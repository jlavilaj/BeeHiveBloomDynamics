```python
import pandas as pd

# Supongamos que tus datos están en un DataFrame llamado df
df = pd.read_csv('DATOS_SERIETEM.csv',  sep=';', low_memory=False)
df['PESO'] = pd.to_numeric(df['PESO'].str.replace(',', '.'), errors='coerce')


# Agrupamos por PERIODO y Horario1a24 y calculamos la media de PESO
mean_weights = df.groupby(['PERIODO', 'Horario1a24'])['PESO'].mean()

# Separamos las medias de peso por periodo
mean_weights_periodo1 = mean_weights[1]
mean_weights_periodo2 = mean_weights[2]
mean_weights_periodo3 = mean_weights[3]

print(mean_weights_periodo1)

print(mean_weights_periodo2)

print(mean_weights_periodo3)
```

Coeficiente de correlación de Pearson. 
Este coeficiente mide la relación lineal entre dos conjuntos de datos. 

Normaliza las series temporales: Resta la media y divide por la desviación estándar de cada serie temporal. Esto te dará dos series temporales con media 0 y desviación estándar 1.
Calcula el producto de las series temporales normalizadas: Multiplica cada punto de datos en la primera serie temporal por el punto de datos correspondiente en la segunda serie temporal.

Suma los productos: Suma todos los productos del paso 2. Esto te dará el numerador del coeficiente de correlación de Pearson.

Divide por el número de puntos de datos: Divide el numerador por el número de puntos de datos en las series temporales para obtener el coeficiente de correlación de Pearson.
El coeficiente de correlación de Pearson estará entre -1 y 1. Un valor de 1 significa que hay una correlación positiva perfecta entre las dos series temporales (cuando una aumenta, la otra también aumenta), un valor de -1 significa que hay una correlación negativa perfecta (cuando una aumenta, la otra disminuye), y un valor de 0 significa que no hay correlación.

Python, puedes usar la función numpy.corrcoef para calcular el coeficiente de correlación de Pearson 


```python
import numpy as np

def pearson(x,y):
    # Calculamos el coeficiente de correlación de Pearson
    correlation_matrix = np.corrcoef(x, y)

    # El coeficiente de correlación es el elemento en la primera fila y la segunda columna
    correlation = correlation_matrix[0, 1]

    return(correlation)
```

```python
import numpy as np


x = mean_weights_periodo1
y = mean_weights_periodo2

correlation = pearson(x,y)

print(f'correlation 1,2: {correlation}')

x = mean_weights_periodo1
y = mean_weights_periodo3

correlation = pearson(x,y)

print(f'correlation 1,3: {correlation}')

x = mean_weights_periodo2
y = mean_weights_periodo3

correlation = pearson(x,y)

print(f'correlation 1,2: {correlation}')
```

prueba de Engle-Granger. Aquí te dejo los pasos para realizar esta prueba:

Comprueba que las series son I(1): Esto significa que las series tienen una raíz unitaria, o en otras palabras, son no estacionarias pero se vuelven estacionarias después de tomar la primera diferencia. Puedes comprobar esto utilizando una prueba de raíz unitaria como la prueba de Dickey-Fuller aumentada.
Estima la relación de largo plazo: Si las series son I(1), estima la relación de largo plazo entre ellas utilizando una regresión ordinaria de mínimos cuadrados.
Comprueba si los residuos son I(0): Los residuos de la regresión del paso 2 representan la desviación a corto plazo de la relación de equilibrio a largo plazo. Si estas series están cointegradas, estos residuos deberían ser I(0), es decir, estacionarios. Puedes comprobar esto utilizando nuevamente una prueba de raíz unitaria.
Si los residuos son estacionarios, entonces puedes concluir que las series están cointegradas.

```python

#NO FUNCIONA VER:
import numpy as np
import statsmodels.api as sm

# Supongamos que tenemos dos series temporales, x e y
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 4, 5, 6])

# Añadimos una constante a x
x = sm.add_constant(x)

# Estimamos la relación de largo plazo
model = sm.OLS(y, x)
results = model.fit()

# Obtenemos los residuos
residuals = results.resid

# Realizamos la prueba de Engle-Granger
eg_test = sm.tsa.stattools.coint(y, x[:, 1], trend='c')

print(eg_test)
```

Los modelos de regresión son una forma común de analizar la relación entre dos o más variables. En el caso de las series temporales, puedes usar un modelo de regresión para predecir una serie temporal en función de otra. Aquí te dejo los pasos para hacerlo:

Preprocesamiento de los datos: Antes de ajustar un modelo de regresión, es posible que necesites preprocesar tus datos. Esto puede incluir la eliminación de tendencias y estacionalidad (por ejemplo, mediante la diferenciación), la normalización de los datos y la comprobación de que los datos son estacionarios.
Divide los datos en conjuntos de entrenamiento y prueba: Es común dividir tus datos en un conjunto de entrenamiento, que utilizarás para ajustar el modelo, y un conjunto de prueba, que utilizarás para evaluar su rendimiento.
Ajusta el modelo de regresión: Utiliza el conjunto de entrenamiento para ajustar un modelo de regresión. Esto podría ser una regresión lineal simple si crees que la relación entre las series temporales es lineal, o podría ser un modelo más complejo como una regresión de soporte vectorial o una red neuronal si crees que la relación es no lineal.
Evalúa el modelo: Utiliza el conjunto de prueba para evaluar el rendimiento del modelo. Esto podría implicar el cálculo de métricas de error como el error cuadrático medio (MSE) o el coeficiente de determinación (R^2).
En Python, puedes usar la biblioteca scikit-learn para ajustar y evaluar modelos de regresión. Aquí tienes un ejemplo de cómo hacerlo con una regresión lineal:

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Supongamos que tenemos dos series temporales, x e y
x = np.array(mean_weights_periodo2).reshape(-1, 1)  # Necesitamos un array 2D para scikit-learn
y = np.array(mean_weights_periodo3)

# Dividimos los datos en conjuntos de entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Ajustamos el modelo de regresión
model = LinearRegression()
model.fit(x_train, y_train)

# Hacemos predicciones en el conjunto de prueba
y_pred = model.predict(x_test)

# Calculamos el error cuadrático medio
mse = mean_squared_error(y_test, y_pred)

print(mse)
```
